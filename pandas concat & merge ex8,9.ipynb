```python
import pandas as pd

Pakistan_weather = pd.DataFrame({
    "city":["karachi","lahore","islamabad"],
    "temperature":[40,34,32],
    "humidity":[85,67,60]
})
Pakistan_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([Pakistan_weather, us_weather])
df
```

```python
df = pd.concat([Pakistan_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([Pakistan_weather, us_weather], keys=["Pakistan","US"])
df
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [76,68,85],
})
df2
```

```python
df3= pd.merge(df1, df2, on="city")
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [28,19,38,42],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [67,72,81],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [28,19,38],
    "humidity": [67,72,81]
})
df2
```

```python
df3= pd.merge (df1, df2, on="city", suffixes =('_left', '_right'))
df3
```
